```python
print(df_cleaned[['duration', 'imdb_score']].describe())
```

```python
# Scatter plot to visualize the relationship between duration and IMDB score
plt.figure(figsize=(12, 6))
sns.scatterplot(x='duration', y='imdb_score', data=df_cleaned, color='coral')
plt.title('Relationship Between Movie Duration and IMDB Score')
plt.xlabel('Duration (minutes)')
plt.ylabel('IMDB Score')
plt.show()
```

```python
# Distribution of movie durations using a histogram
plt.figure(figsize=(12, 6))
sns.histplot(df_cleaned['duration'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Movie Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Regression plot to show the trend
plt.figure(figsize=(12, 6))
sns.regplot(x='duration', y='imdb_score', data=df_cleaned, color='green')
plt.title('Relationship Between Movie Duration and IMDB Score with Regression Line')
plt.xlabel('Duration (minutes)')
plt.ylabel('IMDB Score')
plt.show()
```

```python
language_counts = df_cleaned['language'].value_counts()
```

```python
language_counts
```

```python
# Plot the distribution of languages
plt.figure(figsize=(12, 6))
sns.barplot(x=language_counts.index, y=language_counts.values, palette='viridis')
plt.title('Distribution of Movie Languages')
plt.xlabel('Language')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
language_stats = df_cleaned.groupby('language')['imdb_score'].describe()
```

```python
language_stats
```

```python
# Box plot to visualize the impact of language on IMDB scores
plt.figure(figsize=(14, 8))
sns.boxplot(x='language', y='imdb_score', data=df_cleaned, palette='viridis')
plt.title('Impact of Movie Language on IMDB Scores')
plt.xlabel('Language')
plt.ylabel('IMDB Score')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#D
director_avg_scores = df_cleaned.groupby('director_name')['imdb_score'].mean().sort_values(ascending=False)
```

```python
director_avg_scores
```

```python
# Plot the distribution of IMDB scores for each director
plt.figure(figsize=(20, 6))
sns.boxplot(x='director_name', y='imdb_score', data=df_cleaned, palette='viridis')
plt.title('Distribution of IMDB Scores for Each Director')
plt.xlabel('Director')
plt.ylabel('IMDB Score')
plt.show()
```

```python
# Calculate percentiles for each director's average IMDB score
percentiles = np.percentile(director_avg_scores, [25, 50, 75])

# Display percentiles
print("\nPercentiles of Average IMDB Scores for Directors:")
print(f'25th Percentile: {percentiles[0]}')
print(f'50th Percentile (Median): {percentiles[1]}')
print(f'75th Percentile: {percentiles[2]}')
```

```python
#E
# Scatter plot to visualize the relationship between budget and revenue
plt.figure(figsize=(12, 6))
sns.scatterplot(x='budget', y='gross', data=df_cleaned, color='blue')
plt.title('Relationship Between Movie Budget and Revenue')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.show()
```

```python
correlation_coefficient = df_cleaned['budget'].corr(df_cleaned['gross'])
```

```python
correlation_coefficient
```

```python
# Regression plot to show the trend
plt.figure(figsize=(12, 6))
sns.regplot(x='budget', y='gross', data=df_cleaned, color='green')
plt.title('Relationship Between Movie Budget and Revenue with Regression Line')
plt.xlabel('Budget (Millions)')
plt.ylabel('Revenue (Millions)')
plt.show()
```

```python
df_cleaned['Profit'] = df_cleaned['gross'] - df_cleaned['budget']
df_cleaned['Profit_Margin'] = (df_cleaned['Profit'] / df_cleaned['gross']) * 100
```

```python
df_cleaned[['Profit','Profit_Margin']].info()
```

```python
sorted_movies = df_cleaned.sort_values(by='Profit_Margin', ascending=False)
```

```python
top_10_movies = sorted_movies.head(25)

print("Top 25 Movies with the Highest Profit Margin:")
print(top_10_movies[['movie_title', 'Profit', 'Profit_Margin']])
```
